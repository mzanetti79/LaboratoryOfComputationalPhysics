### 1. Mickey Mouse Problem
a) Write a program that prints the numbers from 1 to 100. But for multiples of three print Mickey instead of the number and for the multiples of five print Mouse. For numbers which are multiples of both three and five print MickeyMouse

b) Put the result in a tuple and substitute Mickey with Donald and Mouse with Duck

```python
# Exercise 1a
List = []
for i in range(1,101) : 
    if ((i%3 == 0) and (i%5 == 0)) : List.append('MickeyMouse')
    elif (i%3 == 0) : List.append('Mickey')
    elif (i%5 == 0) : List.append('Mouse')
    else : List.append(i)
        
for item in List : print (item)

# Exercise 1b    
OriginalTuple = tuple(List)

for i,item in enumerate(List) : 
    if item == 'Mickey' : List[i] = 'Donald'
    elif item == 'Mouse' : List[i] = 'Duck'
        
NewTuple = tuple(List)
print (NewTuple)
```

### 3. Computing the distance

Write a function that calculates and returns the euclidean distance between two points u* and *v, where u* and *v are both 2-tuples (x,y). For example, if u=(3,0) and v=(0,4), the function should return 5


```python
# Exercise 3
import math

def Distance(u,v) :
    dx = abs(u[0] - v[0])
    dy = abs(u[1] - v[1])
    distance = math.sqrt(dx**2 + dy**2)
    return distance

# Test our function
u,v = (3,0),(0,4)
print (Distance(u,v))
```

### 4. Counting letters

Write a program to calculate the number of times each character occurs in a given string s. Ignore differneces in capitalization


```python
# Exercise 4
import string

s="Write a program that prints the numbers from 1 to 100. \
But for multiples of three print Mickey instead of the number and for the multiples of five print Mouse. \
For numbers which are multiples of both three and five print MickeyMouse"

def CountLetters(s) : 
    s = s.lower()
    for letter in string.ascii_lowercase : 
        n = s.count(letter)
        print (letter + ' Has been found ' + str(n) + ' times!')
        
    return

# Test function
CountLetters(s)
```

### 6. Combination of functions

Write two functions - one that returns the square of a number, and one that returns the cube. Now write a third function that returns the number raised to the 6th power using the two previous functions.


```python
# Exercise 6

def square(f) :
    return f**2

def cube(f) :
    return f**3

def PowerSix(f) :
    return cube(square(f))

# Test function
print (PowerSix(2))
```

```python
# Exercise 8 - Find Pythagorean Triples

Range = 100
Pyth_List = []
for b in range(1,Range) :
    for a in range(1,b) :
        c = math.sqrt(a**2 + b**2)
        if ((c%1 == 0) and (c < 100)) : 
            PTuple = (a,b,int(c))
            Pyth_List.append(PTuple)
            
for item in Pyth_List : print (item)
```

### 9. Normalization

Write a function that takes a tuple of numbers and returns it with the entries normalized to one


```python
# Exercise 9 - Normalization

def Normalization(t) :
    N = 0.0
    l = list(t)
    for i in range(0,len(t)) :
        N += t[i]**2
    for i in range(0,len(l)) :
        l[i] = l[i]/math.sqrt(N)
    return tuple(l)

# Test function
t1 = (0,3,4,0)
t2 = (-4,12,8,-2)
n1, n2 = Normalization(t1), Normalization(t2)
print (n1)
print (n2)
```
