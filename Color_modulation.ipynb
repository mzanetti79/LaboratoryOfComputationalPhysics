```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def convert_from_10(stringa, base):
    num = int(stringa)
    l = ""
    while num >= base:
        resto = num%base
        num = int((num - resto)/base)
        if resto == 10:
            resto = "A"
        if resto == 11:
            resto = "B"
        if resto == 12:
            resto = "C"
        if resto == 13:
            resto = "D"
        if resto == 14:
            resto = "E"
        if resto == 15:
            resto = "F"
        l = str(resto) + l


    resto = num%base

    if resto == 10:
         resto = "A"
    if resto == 11:
        resto = "B"
    if resto == 12:
        resto = "C"
    if resto == 13:
        resto = "D"
    if resto == 14:
        resto = "E"
    if resto == 15:
        resto = "F"

    l = str(resto) + l

    return l

def make_color(hexadec = 0):
    num = int(hexadec)
    l = ""
    base = 16                 # C' una base generica: l'ho copiato dal convertitore di numeri in base qualsiasi
    while num >= base:
        resto = num%base
        num = int((num - resto)/base)
        if resto == 10:
            resto = "A"
        if resto == 11:
            resto = "B"
        if resto == 12:
            resto = "C"
        if resto == 13:
            resto = "D"
        if resto == 14:
            resto = "E"
        if resto == 15:
            resto = "F"
        l = str(resto) + l


    resto = num%base

    if resto == 10:
         resto = "A"
    if resto == 11:
        resto = "B"
    if resto == 12:
        resto = "C"
    if resto == 13:
        resto = "D"
    if resto == 14:
        resto = "E"
    if resto == 15:
        resto = "F"
    l = str(resto) + l
    
    while len(l)<6:
        l = '0'+l
    
    l = '#'+l
    
    return l


fig, ax = plt.subplots ( nrows = 1, ncols = 1, figsize = (20, 10))

x = np.arange(100)

# Trying color modulation
col_min = 0x000000     # 0x000000 (r = 00,       g = 00,       b = 00) = black
col_max = 0xffff00     # 0xffff00 (r = ff (max), g = ff (max), b = 00) = yellow

for cont in range(len(x)):
    color_ = (col_min + int(cont/float(len(x)-1)*col_max))
    color_ = make_color((str(color_)).replace('0x', '#'))
    print(cont, '\t', color_)
    ax.plot(x[cont], x[cont], marker = 's', markersize = 20, markeredgecolor = color_, markerfacecolor = color_, label=cont )
    
ax.legend()

#print(interv, x)
```
